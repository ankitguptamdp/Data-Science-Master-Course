# 01. Data Preprocessing - Normalisation or Standardisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfx = pd.read_csv('../Csv Files/linearX.csv')
dfy = pd.read_csv('../Csv Files/linearY.csv')
x = dfx.values
y = dfy.values
x2 = (x-x.mean())/x.std()
plt.scatter(x,y,color='yellow',label='Raw Data')
plt.scatter(x2,y,c='red',label='Normalised Data')
plt.legend()
plt.show()
```
