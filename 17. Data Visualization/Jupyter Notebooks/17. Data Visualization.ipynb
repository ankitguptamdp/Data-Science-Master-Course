# 07. Movie Data Visualization

### Data Visualisation Challenge
Given a Movie Dataset, your task is to make a visualization that
- Plots the Length of Movie Title Names on X-Axis
- Frequency Counts on Y-Axis i.e., Number of Movies having 'x' characters in their title

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read a CSV File
df = pd.read_csv("../Csv Files/movie_metadata.csv")
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles = df.get('movie_title')
```

```python
print(titles)
```

```python
print(type(titles))
```

```python
titles = list(df.get('movie_title'))
print(titles[:5])
print(type(titles))
```

```python
print(titles[0])
```

```python
freq_titles = {}

for title in titles:
    length = len(title)
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
print(X)
print(Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Freq Count")
plt.title("Movie Data Visualization Problem")
plt.show()
```
